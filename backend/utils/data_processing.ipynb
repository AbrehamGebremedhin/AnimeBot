```python
# Find the most common value in each column
most_common_values = data.mode().iloc[0]

# Find the number of occurrences of each most common value
occurrences = {col: data[col].value_counts().get(most_common_values[col], 0) for col in data.columns}

# Display the most common values and their occurrences
for col in data.columns:
    print(f"Column: {col}, Most Common Value: '{most_common_values[col]}', Occurrences: {occurrences[col]}")
```
